```python
from IPython.core.display import Image
```

**Set the GRASS working region**

```python
!g.region res=1 -ap save=bathy --o
```

**Import and display digital bathymetry**

[r.in.gdal](https://grass.osgeo.org/grass71/manuals/r.in.gdal.html)
[d.mon](https://grass.osgeo.org/grass71/manuals/d.mon.html)

```python
!r.in.gdal input=DATA/bathy.tif output=bathy -o --o
!r.null map=bathy setnull=-9999 
!r.colors -e map=bathy color=haxby 
```

```python
layer='bathy'
!d.mon start=cairo --q output=TMP/{layer}.png --overwrite
!d.rast map={layer} --q
!d.mon stop=cairo --q
Image('TMP/'+layer+'.png')
```
